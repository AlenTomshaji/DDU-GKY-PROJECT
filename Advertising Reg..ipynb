```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')
import klib as k
```

```python
data=pd.read_csv('advertising.csv')
data.head(10)
```

```python
data.isnull().sum()
```

```python
k.corr_mat(data)
```

```python
k.corr_plot(data=data, target='Sales');
```

```python
for i in data.columns:
    if data[i].dtype !="object":
        sns.boxplot(y=data[i], color="red")
        plt.show();
```

```python
x=data[['TV','Radio']].values
x
```

```python
y=data[['Sales']].values
sc=StandardScaler()
x=sc.fit_transform(x)
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
training_score=model.score(x_train,y_train)*100
training_score
```

```python
testing_score=model.score(x_test,y_test)*100
testing_score
```

```python
R2_score= r2_score(y_test,model.predict(x_test))*100
R2_score
```

```python
mae = mean_absolute_error(y_test,model.predict(x_test))
mae
```

```python
mse = mean_squared_error(y_test,model.predict(x_test))
mse
```

```python
y_pred =model.predict(x)
y_pred
```

```python
out=pd.DataFrame(zip(y,y_pred), columns=("ACTUAL","PREDICTED"),dtype=float)
out.head()
```

```python
plt.figure(figsize=(12,6))
plt.title("PREDICTED VS ACTUAL")
sns.lineplot(data=out, x="ACTUAL", y = "PREDICTED");
```
